```python
# RNA Splicing

"""
Pipeline:
1. see file = view whole file
2. Parse file = idenfiy first rosalind marker
3. seperate = split gene introns
4. splice parts = splice introns
5. Transcribe = Replace thymine with uracil
6. Translate = Convert codons to amino acids
"""
```

```python
def see_file(filepath):
    with open(filepath) as f:
          data = f.read().strip()
    return data

def parse_file(data):
    sequences = []
    current_sequence = ""

    for line in data.splitlines():
        if line.startswith(">"):
            if current_sequence:
                sequences.append(current_sequence)

            current_sequence = ""

        else:
            current_sequence += line.strip()

    if current_sequence:
        sequences.append(current_sequence)

    return sequences

def seperate(sequence):
    whole = sequence[0]
    parts = sequence[1:]

    return whole, parts

# Whole referse to the whole sequence, parts refers to the introns
def splice_parts(whole, parts):

    for part in parts:
        whole = whole.replace(part, "")

    return whole

def transcribe(whole):
    return whole.replace("T", "U")

codon_dict = {
    "UUU": "F", "UUC": "F",
    "UUA": "L", "UUG": "L", "CUU": "L", "CUC": "L", "CUA": "L", "CUG": "L",
    "AUU": "I", "AUC": "I", "AUA": "I",
    "AUG": "M", # the start codon
    "GUU": "V", "GUC": "V", "GUA": "V", "GUG": "V",
    "UCU": "S", "UCC": "S", "UCA": "S", "UCG": "S", "AGU": "S", "AGC": "S",
    "CCU": "P", "CCC": "P", "CCA": "P", "CCG": "P",
    "ACU": "T", "ACC": "T", "ACA": "T", "ACG": "T",
    "GCU": "A", "GCC": "A", "GCA": "A", "GCG": "A",
    "UAU": "Y", "UAC": "Y",
    "CAU": "H", "CAC": "H",
    "CAA": "Q", "CAG": "Q",
    "AAU": "N", "AAC": "N",
    "AAA": "K", "AAG": "K",
    "GAU": "D", "GAC": "D",
    "GAA": "E", "GAG": "E",
    "UGU": "C", "UGC": "C",
    "UAA": " ", "UAG": " ", "UGA": " ", # the stop codons
    "UGG": "W",
    "CGU": "R", "CGC": "R", "CGA": "R", "CGG": "R", "AGA": "R", "AGG": "R",
    "GGU": "G", "GGC": "G", "GGA": "G", "GGG": "G",
}

def translate(whole, codon_dict):
    amino_acids = []
    
    for nuc in range(0, len(whole), 3):
        
        codon = whole[nuc:nuc + 3]
        
        if len(codon) < 3:
            break 

        if codon in ["UAA", "UAG", "UGA"]:
            break

        amino_acids.append(codon_dict[codon])

    return "".join(amino_acids)
    
# ALL print statements

# Change User_name to your User name
filepath = r"C:\Users\User_name\Downloads\rosalind_splc.txt"

data = see_file(filepath)
#print(f"View the file: {data}") # uncomment to view the complete file

sequence = parse_file(data)
#print(f"View all the Sequence: {sequence}") # uncomment to see all sequences

whole, parts = seperate(sequence)
whole = splice_parts(whole, parts)
#print(f"View the sequence with introns removed: {whole}") # uncomment to view the whole "sequence" with parts "Introns" removed

whole = transcribe(whole)
#print(f"View Transcribed sequence: {whole}") # uncommit to view transcribed sequence 

results = translate(whole, codon_dict)
print(f"View peptide sequence with introns removed: {results}")
```
