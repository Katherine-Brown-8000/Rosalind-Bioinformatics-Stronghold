```python
# Locating Restriction site

"""
Pipeline:
1. Read FASTA file
2. Create reverse complement function
Generate fragments (lengths 4–12)
Compare each fragment to its reverse complement
Output position and length of matching palindromes
"""
```

```python
def read_file(filepath):
    with open(filepath) as f:
        lines = f.readlines()

        sequence = ""

        for line in lines:
            if not line.startswith(">"):
                sequence += line.strip()

        return sequence

def reverse_complement(sequence):
    reverse_seq = sequence[::-1]
    complement = {"A": "T","T": "A","C": "G","G":"C"}
    reverse_comp = ""

    for nuc in reverse_seq:
        reverse_comp += complement[nuc]
        
    return reverse_comp

def find_restriction_sites(sequence):
    results = []
    for start in range(len(sequence)):
        for length in [4, 6, 8, 10, 12]:
            fragment = sequence[start:start+length]
            if len(fragment) == length:
                if fragment == reverse_complement(fragment):
                    position = start + 1
                    results.append((position, length))

    return results
            
    
# Change User_name to your User name
filepath = r"C:\Users\scien\Downloads\rosalind_revp.txt"

sequence = read_file(filepath)
#print(sequence) # Uncommit to view sequence from rosalind file

reverse_comp = reverse_complement(sequence)
#print(reverse_comp) # Uncommit to view reverse complement

results = find_restriction_sites(sequence)
for position, length in results:
    print(position, length)

#print(results)
```
