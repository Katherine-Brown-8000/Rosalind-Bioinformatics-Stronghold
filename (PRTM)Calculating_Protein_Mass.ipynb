```python
# Calculating Protein Mass
"""
Calculate the weight of a peptide based off of a monoisotopic mass table.

Pipeline:
1. Reference Data - Monoisotopic mass table
2. Read input - The rosalind_prtm file
3. Computation -  Calculate peptide weight
    3a. Mapping - converting amimo acids to masses
    3b. Reduction - Sum all masses
"""
```

```python
monoisotopic_mass_table = {
    "A": 71.03711,
    "R": 156.10111,
    "N": 114.04293,
    "D": 115.02694,
    "C": 103.00919,
    "E": 129.04259,
    "Q": 128.05858,
    "G": 57.02146,
    "H": 137.05891,
    "I": 113.08406,
    "L": 113.08406,
    "K": 128.09496,
    "M": 131.04049,
    "F": 147.06841,
    "P": 97.05276,
    "S": 87.03203,
    "T": 101.04768,
    "W": 186.07931,
    "Y": 163.06333,
    "V": 99.06841
}

def calculate_weight_peptide(filepath):
    # Read rosalind's amino acid sequence
    with open(filepath) as f:
        protein_info = f.read().strip()

    total_mass = 0.0

    for aa in protein_info:
        total_mass += monoisotopic_mass_table[aa]

    return total_mass

# Change User_name to your User name
filepath = r"C:\Users\User_name\Downloads\rosalind_prtm.txt"
result = calculate_weight_peptide(filepath)
print(result)
```
